```python
print(torch.cuda.is_available()) #false MT running on cpu
text = "hi, how are you this is hemanth from chennai"
input_id = tokenizer([text],return_tensors="pt")
```

```python
input_id
```

```python
%%time
tar_gen=model.generate(**input_id)
```

```python
res=tokenizer.batch_decode(tar_gen, skip_special_tokens=True)[0]
print(res)
```
